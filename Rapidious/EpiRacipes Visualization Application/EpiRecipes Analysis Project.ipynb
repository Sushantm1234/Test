##### **Missing or Null Values**

```python
# checking missing or null values in dataset
epirecipes.isnull().sum()
```

##### **Handle Missing data**

```python
# filling missing values with median
epirecipes['calories'] = epirecipes['calories'].fillna(epirecipes['calories'].median())
epirecipes['protein'] = epirecipes['protein'].fillna(epirecipes['protein'].median())
epirecipes['fat'] = epirecipes['fat'].fillna(epirecipes['fat'].median())
epirecipes['sodium'] = epirecipes['sodium'].fillna(epirecipes['sodium'].median())
```

```python
# verifying null or missing values
epirecipes.isnull().sum()
```

##### **Duplicated values**

```python
# checking duplicated values in dataset
epirecipes.duplicated().sum()
```

```python
# all rows and all columns
epirecipes.shape
```

```python
# droping duplicated values 
epirecipes = epirecipes.drop_duplicates()
```

```python
# Drop rows with missing ratings where ratings are necessary for analysis
epirecipes = epirecipes.dropna(subset=['rating'])
```

#### **Data Visualization**

##### **Q.1 - What are the most common ingredients in highly rated recipes?**

##### Top 10 Most Common Ingredients in highly-rated recipes

```python
# Filter recipes with rating >= 4.5 (highly rated)
highly_rated = epirecipes[epirecipes['rating'] >= 4.5]

# Exploding ingredient-related columns for analysis
ingredient_columns = [col for col in epirecipes.columns if col not in ['title', 'rating', 'calories', 'protein', 'fat', 'sodium']]

# Summing binary ingredient columns
ingredient_sums = highly_rated[ingredient_columns].sum().sort_values(ascending=False).head(10)

# Plotting top 10 ingredients
sns.barplot(x=ingredient_sums.values, y=ingredient_sums.index)
plt.title('Top 10 Ingredients in Highly Rated Recipes')
plt.xlabel('Count')
plt.ylabel('Ingredients')
plt.show()
```

##### **Observations** -  From the above graph, bon appetit is the top most common ingredients in highly rated recipes. also the peanut free, soy free, tree nut free, vegetarian, kosher, pescatarian, wheat/gluten free, gourmet, summer are also most common ingredients in highly rated recipes.

##### **Q.2 - Are there correlations between preparation time and recipe ratings?**

##### Correlations between preparation time and recipe ratings

```python
epirecipes.head()
```

###### *In the dataset, there is no preparation time column but '22-minute meals' column is there, which contains the values as 0 means 'No' and 1 means 'Yes'.*

```python
# Relation between 22-minute meals and rating
sns.scatterplot(x='22-minute meals', y='rating', data=epirecipes)
plt.title('22-minute meals vs Rating')
plt.xticks([0, 1])
plt.show()
```

##### **Observations** - From the above graph we can see that, The meals which are not takes to prepare 22 minutes having more number of count of ratings than meals which takes 22 minutes time.

```python
# Calculating correlation between '22-minute meals' and 'rating'
correlation = epirecipes[['22-minute meals', 'rating']].corr()

# Print the correlation matrix
print(correlation)
```

##### **Observations** - From the above, we can see that there is very less i.e. 0.011763 means 1.1763% correlation between '22-minute meals' and 'rating'. So we can say that there is No correlation between '22-minute meals' and 'rating'.
